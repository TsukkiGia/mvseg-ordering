# Procedure policy mean/CI bar-grid

This notebook loads Plan B outputs for a given **procedure** and plots a **grid of dataset bar charts**.

For each dataset (family):
- x-axis: policy
- y-axis: hierarchical bootstrap mean of selected metric
- error bars: hierarchical bootstrap CI

You can change the metric to `initial_dice`, `iterations_used`, or `final_dice` (or any metric column in Plan B CSVs).

```python
from __future__ import annotations

import math
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sys

from matplotlib.ticker import FormatStrFormatter, MultipleLocator

REPO_ROOT = Path('/data/ddmg/mvseg-ordering')
for path in [REPO_ROOT, REPO_ROOT / 'UniverSeg', REPO_ROOT / 'MultiverSeg', REPO_ROOT / 'ScribblePrompt']:
    if str(path) not in sys.path:
        sys.path.append(str(path))

from experiments.analysis.hierarchical_ci import (
    compute_subset_scores,
    dataset_bootstrap_stats,
    hierarchical_bootstrap_dataset,
    hierarchical_bootstrap_task_estimates,
)
from experiments.analysis.task_explorer import FAMILY_ROOTS
```

```python
# -----------------------------
# Config
# -----------------------------
REPO_ROOT = Path('/data/ddmg/mvseg-ordering')
PROCEDURE = 'random_vs_uncertainty_final'
METRICS = ['initial_dice', 'iterations_used', 'final_dice']

METRIC_DISPLAY = {
    'initial_dice': 'Initial Dice Score',
    'final_dice': 'Final Dice Score',
    'iterations_used': 'Interactions Used',
}

# Optional ablation filter. If None, notebook auto-picks when a single ablation exists,
# or falls back to the first ablation alphabetically.
ABLATION = None

# Keep only these policies (in this order).
POLICY_ORDER = ['random', 'reverse_curriculum', 'curriculum']
POLICY_DISPLAY_NAMES = {
    'random': 'random',
    'reverse_curriculum': 'easy → hard',
    'curriculum': 'hard → easy',
}

# Paired-difference mode uses subset-level pairing against baseline policy,
# then hierarchical bootstrap over paired deltas at task->dataset levels.
DELTA_VS_RANDOM = True
BASELINE_POLICY = 'random'
INCLUDE_BASELINE_IN_PLOT = False
FAMILY_DISPLAY = {'T1mix': 'COBRE'}

N_BOOT = 2000
SEED = 23
N_COLS = 3

SAVE_FIG = False
FIG_DIR = REPO_ROOT / 'figures' / 'procedure_policy_bar_grid'

# View controls
ZOOM_MODE = 'local'       # 'none', 'local', or 'global'
ZOOM_PAD_FRAC = 0.20
from experiments.analysis.notebook_plot_style import (
    FONT_SIZE_SUPTITLE,
    FONT_SIZE_SUPYLABEL,
    FONT_SIZE_PANEL_TITLE,
    FONT_SIZE_XTICKS,
    FONT_SIZE_YTICKS,
    BAR_Y_TICK_STEP_DICE,
    BAR_Y_TICK_STEP_ITERS,
)
```

```python
def load_planb_for_procedure(
    repo_root: Path,
    procedure: str,
    *,
    filename: str = 'subset_support_images_summary.csv',
) -> pd.DataFrame:
    # Load all Plan B summaries for all families/tasks/ablations/policies in one procedure.
    scripts_root = repo_root / 'experiments' / 'scripts' / procedure
    frames: list[pd.DataFrame] = []

    for root_name, family in FAMILY_ROOTS.items():
        family_root = scripts_root / root_name
        if not family_root.exists():
            continue

        for task_dir in sorted(p for p in family_root.iterdir() if p.is_dir()):
            task_id = f'{family}/{task_dir.name}'

            for ablation_dir in sorted(p for p in task_dir.iterdir() if p.is_dir()):
                # Preferred layout: <ablation>/<policy>/B/subset_support_images_summary.csv
                policy_dirs = sorted(
                    p for p in ablation_dir.iterdir()
                    if p.is_dir() and (p / 'B' / filename).exists()
                )

                if policy_dirs:
                    for policy_dir in policy_dirs:
                        csv_path = policy_dir / 'B' / filename
                        df = pd.read_csv(csv_path)
                        if df.empty:
                            continue
                        df['family'] = family
                        df['task_id'] = task_id
                        df['task_name'] = task_dir.name
                        df['ablation'] = ablation_dir.name
                        df['policy_name'] = policy_dir.name
                        df['__source__'] = str(csv_path)
                        frames.append(df)
                    continue

                # Fallback layout: <ablation>/B/subset_support_images_summary.csv
                csv_path = ablation_dir / 'B' / filename
                if csv_path.exists():
                    df = pd.read_csv(csv_path)
                    if df.empty:
                        continue
                    df['family'] = family
                    df['task_id'] = task_id
                    df['task_name'] = task_dir.name
                    df['ablation'] = ablation_dir.name
                    df['policy_name'] = 'random'
                    df['__source__'] = str(csv_path)
                    frames.append(df)

    if not frames:
        raise FileNotFoundError(
            f'No Plan B summaries found under experiments/scripts/{procedure}/**/{filename}'
        )

    return pd.concat(frames, ignore_index=True)


def policy_sort_key(name: str) -> tuple[int, str]:
    n = str(name)
    try:
        return (POLICY_ORDER.index(n), n)
    except ValueError:
        return (len(POLICY_ORDER) + 1, n)
```

```python
raw_df = load_planb_for_procedure(REPO_ROOT, PROCEDURE)
print(f'Loaded rows: {len(raw_df):,}')
print('Families:', sorted(raw_df['family'].unique().tolist()))
print('Ablations:', sorted(raw_df['ablation'].unique().tolist()))

metrics_to_run = [m for m in METRICS if m in raw_df.columns]
missing_metrics = [m for m in METRICS if m not in raw_df.columns]
if not metrics_to_run:
    raise ValueError(f'None of METRICS are available. Requested={METRICS}')
if missing_metrics:
    print(f'[warn] Missing metrics skipped: {missing_metrics}')

available_ablations = sorted(raw_df['ablation'].astype(str).unique().tolist())
if ABLATION is None:
    if len(available_ablations) == 1:
        active_ablation = available_ablations[0]
    else:
        active_ablation = available_ablations[0]
        print(f"[info] Multiple ablations found: {available_ablations}")
        print(f"[info] Using first by default: {active_ablation}. Set ABLATION to override.")
else:
    active_ablation = str(ABLATION)
    if active_ablation not in available_ablations:
        raise ValueError(f"ABLATION={active_ablation!r} not in available ablations: {available_ablations}")

df = raw_df[raw_df['ablation'].astype(str) == active_ablation].copy()
df = df[df['policy_name'].astype(str).isin(POLICY_ORDER)].copy()

if df.empty:
    raise ValueError(f"No rows left after filtering to POLICY_ORDER={POLICY_ORDER}")

print(f'Using ablation: {active_ablation}')
print('Families in ablation:', sorted(df['family'].unique().tolist()))
print('Policies kept:', sorted(df['policy_name'].astype(str).unique().tolist(), key=policy_sort_key))
print('Metrics to run:', metrics_to_run)
```

```python
def build_summary_for_metric(metric_name: str) -> pd.DataFrame:
    rows: list[dict[str, object]] = []

    for family, fam_df in df.groupby('family'):
        available = set(fam_df['policy_name'].astype(str).tolist())
        policies = [p for p in POLICY_ORDER if p in available]

        if DELTA_VS_RANDOM:
            if BASELINE_POLICY not in policies:
                print(f"[warn] family={family}: missing baseline policy={BASELINE_POLICY}; skipping")
                continue

            subset_scores = compute_subset_scores(
                fam_df,
                metric_name,
                task_col='task_id',
                policy_col='policy_name',
            )

            baseline_subset = (
                subset_scores[subset_scores['policy_name'].astype(str) == BASELINE_POLICY]
                [['task_id', 'subset_index', 'subset_mean']]
                .rename(columns={'subset_mean': 'baseline_subset_mean'})
            )

            # Baseline row is exactly zero delta by definition.
            rows.append(
                {
                    'family': str(family),
                    'ablation': active_ablation,
                    'metric': metric_name,
                    'policy_name': BASELINE_POLICY,
                    'mean': 0.0,
                    'ci_lo': 0.0,
                    'ci_hi': 0.0,
                    'n_tasks': int(baseline_subset['task_id'].nunique()),
                    'n_subsets': int(len(baseline_subset)),
                }
            )

            for policy_name in policies:
                if policy_name == BASELINE_POLICY:
                    continue

                policy_subset = (
                    subset_scores[subset_scores['policy_name'].astype(str) == policy_name]
                    [['task_id', 'subset_index', 'subset_mean']]
                    .rename(columns={'subset_mean': 'policy_subset_mean'})
                )

                paired = policy_subset.merge(
                    baseline_subset,
                    on=['task_id', 'subset_index'],
                    how='inner',
                    validate='one_to_one',
                )

                if paired.empty:
                    print(f"[warn] family={family}, policy={policy_name}: no paired subsets with baseline")
                    continue

                paired['delta'] = paired['policy_subset_mean'] - paired['baseline_subset_mean']

                subset_scores_by_task = {
                    str(task): grp['delta'].to_numpy(dtype=float)
                    for task, grp in paired.groupby('task_id')
                }

                task_boot = hierarchical_bootstrap_task_estimates(
                    subset_scores_by_task,
                    n_boot=N_BOOT,
                    seed=SEED,
                )
                _, mean, lo, hi = dataset_bootstrap_stats(task_boot, alpha=0.05)

                rows.append(
                    {
                        'family': str(family),
                        'ablation': active_ablation,
                        'metric': metric_name,
                        'policy_name': str(policy_name),
                        'mean': float(mean),
                        'ci_lo': float(lo),
                        'ci_hi': float(hi),
                        'n_tasks': int(paired['task_id'].nunique()),
                        'n_subsets': int(len(paired)),
                    }
                )
        else:
            for policy_name in policies:
                stats = hierarchical_bootstrap_dataset(
                    fam_df,
                    metric_name,
                    policy_name,
                    n_boot=N_BOOT,
                    seed=SEED,
                    task_col='task_id',
                    policy_col='policy_name',
                )
                rows.append(
                    {
                        'family': str(family),
                        'ablation': active_ablation,
                        'metric': metric_name,
                        'policy_name': str(policy_name),
                        'mean': float(stats['mean']),
                        'ci_lo': float(stats['ci_lo']),
                        'ci_hi': float(stats['ci_hi']),
                        'n_tasks': int(fam_df['task_id'].nunique()),
                        'n_subsets': int(fam_df[['task_id', 'subset_index']].drop_duplicates().shape[0]),
                    }
                )

    metric_summary = pd.DataFrame(rows)
    if metric_summary.empty:
        return metric_summary

    metric_summary['policy_name'] = pd.Categorical(
        metric_summary['policy_name'],
        categories=POLICY_ORDER,
        ordered=True,
    )
    metric_summary = metric_summary.sort_values(['family', 'policy_name']).reset_index(drop=True)
    metric_summary['policy_name'] = metric_summary['policy_name'].astype(str)
    return metric_summary


summary_by_metric: dict[str, pd.DataFrame] = {}
for metric_name in metrics_to_run:
    metric_summary = build_summary_for_metric(metric_name)
    if metric_summary.empty:
        print(f"[warn] No summary rows for metric={metric_name}")
        continue
    summary_by_metric[metric_name] = metric_summary

if not summary_by_metric:
    raise ValueError('No summary rows produced for any requested metric.')

summary_df = pd.concat(summary_by_metric.values(), ignore_index=True)
summary_df
```

```python
def plot_dataset_bar_grid(summary_df: pd.DataFrame) -> None:
    if summary_df.empty:
        print('No summary rows to plot.')
        return

    zoom_mode = str(ZOOM_MODE).strip().lower()
    if zoom_mode not in {'none', 'local', 'global'}:
        raise ValueError(f"Unsupported ZOOM_MODE={ZOOM_MODE!r}. Use 'none', 'local', or 'global'.")

    metric_name = str(summary_df['metric'].iloc[0])
    metric_label = METRIC_DISPLAY.get(metric_name, metric_name.replace('_', ' ').title())
    y_label = f'Δ {metric_label} vs. Random' if DELTA_VS_RANDOM else metric_label

    families = sorted(summary_df['family'].unique().tolist())
    policy_order = [p for p in POLICY_ORDER if p in set(summary_df['policy_name'].astype(str).tolist())]
    if not INCLUDE_BASELINE_IN_PLOT and BASELINE_POLICY in policy_order:
        policy_order = [p for p in policy_order if p != BASELINE_POLICY]
    if not policy_order:
        raise ValueError('No policies left to plot after applying INCLUDE_BASELINE_IN_PLOT filter.')

    # Consistent policy colors.
    policy_colors = {
        'random': '#4C78A8',
        'reverse_curriculum': '#F58518',
        'curriculum': '#54A24B',
    }

    n_families = len(families)
    n_cols = 3
    n_rows = 4

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(4.0 * n_cols, 3.8 * n_rows), squeeze=False, sharey=True)
    axes_flat = axes.ravel()

    global_lo = float(summary_df['ci_lo'].min())
    global_hi = float(summary_df['ci_hi'].max())
    global_span = max(global_hi - global_lo, 1e-8)
    global_pad = ZOOM_PAD_FRAC * global_span

    for i, family in enumerate(families):
        ax = axes_flat[i]
        ax.tick_params(axis='y', labelleft=True)
        tick_cfg = {
            'iterations_used': (BAR_Y_TICK_STEP_ITERS, '%.1f'),
            'initial_dice': (BAR_Y_TICK_STEP_DICE, '%.2f'),
            'final_dice': (BAR_Y_TICK_STEP_DICE, '%.2f'),
        }
        tick_step, tick_fmt = tick_cfg.get(metric_name, (BAR_Y_TICK_STEP_DICE, '%.2f'))
        if tick_step is not None:
            ax.yaxis.set_major_locator(MultipleLocator(tick_step))
            ax.yaxis.set_major_formatter(FormatStrFormatter(tick_fmt))
        fam = summary_df[summary_df['family'] == family].copy()
        fam = fam[fam['policy_name'].astype(str).isin(policy_order)].copy()
        fam['policy_name'] = pd.Categorical(fam['policy_name'], categories=policy_order, ordered=True)
        fam = fam.sort_values('policy_name').reset_index(drop=True)
        if fam.empty:
            ax.axis('off')
            continue

        x = np.arange(len(fam), dtype=float)
        means = fam['mean'].to_numpy(dtype=float)
        ci_lo = fam['ci_lo'].to_numpy(dtype=float)
        ci_hi = fam['ci_hi'].to_numpy(dtype=float)
        err_lo = means - ci_lo
        err_hi = ci_hi - means

        colors = [policy_colors.get(str(p), '#777777') for p in fam['policy_name'].tolist()]
        ax.bar(x, means, color=colors, alpha=0.85, width=0.72)
        ax.errorbar(x, means, yerr=np.vstack([err_lo, err_hi]), fmt='none', ecolor='black', capsize=3, linewidth=1)

        if DELTA_VS_RANDOM:
            ax.axhline(0.0, color='black', linestyle='--', linewidth=1.0, alpha=0.8)

        # if zoom_mode == 'local':
        #     lo = float(ci_lo.min())
        #     hi = float(ci_hi.max())
        #     span = max(hi - lo, 1e-8)
        #     pad = ZOOM_PAD_FRAC * span
        #     ax.set_ylim(lo - pad, hi + pad)
        # elif zoom_mode == 'global':
        ax.set_ylim(global_lo - global_pad, global_hi + global_pad)

        ax.set_title(FAMILY_DISPLAY.get(family, family), fontsize=FONT_SIZE_PANEL_TITLE)
        ax.set_xticks(x)
        ax.tick_params(axis='y', labelsize=FONT_SIZE_YTICKS)
        ax.set_xticklabels(
            [POLICY_DISPLAY_NAMES.get(str(p), str(p)) for p in fam['policy_name'].tolist()],
            rotation=25,
            ha='right',
            fontsize=FONT_SIZE_XTICKS,
        )
        ax.grid(axis='y', linestyle='--', alpha=0.35)

    for j in range(n_families, len(axes_flat)):
        axes_flat[j].axis('off')

    if DELTA_VS_RANDOM:
        title = (
            f"Paired delta of uncertainty vs random on {metric_label.lower()}: "
            f"Average starts across 10 runs"
        )
    else:
        title = f'{metric_label} by Ordering Policy (Hierarchical Bootstrap 95% CI)'

    fig.suptitle(title, fontsize=FONT_SIZE_SUPTITLE, y=0.995)
    fig.supylabel(y_label, fontsize=FONT_SIZE_SUPYLABEL, x=0.02)
    fig.tight_layout(rect=(0.03, 0, 1, 0.97))

    if SAVE_FIG:
        FIG_DIR.mkdir(parents=True, exist_ok=True)
        mode_tag = 'paired_delta' if DELTA_VS_RANDOM else 'absolute'
        zoom_tag = zoom_mode
        ablation_name = str(summary_df['ablation'].iloc[0])
        out_path = FIG_DIR / f"{PROCEDURE.replace('/', '_')}_{ablation_name}_{metric_name}_{mode_tag}_{zoom_tag}_bar_grid.png"
        fig.savefig(out_path, dpi=180, bbox_inches='tight')
        print(f'saved: {out_path}')

    plt.show()


for metric_name in metrics_to_run:
    metric_summary = summary_by_metric.get(metric_name)
    if metric_summary is None or metric_summary.empty:
        continue
    plot_dataset_bar_grid(metric_summary)
```
